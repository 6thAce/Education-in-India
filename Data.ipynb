```python
import pandas as pd

district = pd.read_csv("Data set/2015_16_Districtwise.csv")
elementary = pd.read_csv("Data set/2015_16_Statewise_Elementary.csv")
secondary = pd.read_csv("Data set/2015_16_Statewise_Secondary.csv")

district.head()
```

```python
district = district.dropna(axis=1, how='all')
district = district.dropna(how='all')
```

```python
district["STATNAME"] = district["STATNAME"].str.strip().str.title()
district["DISTNAME"] = district["DISTNAME"].str.strip().str.title()
```

```python
district = district.dropna(subset=["STATNAME", "DISTNAME"])
```

```python
text_cols = ["STATNAME", "DISTNAME"]

for col in district.columns:
    if col not in text_cols:
        district[col] = pd.to_numeric(district[col], errors='coerce')
```

```python
district = district.fillna(0)
```

```python
district = district.drop_duplicates()
```

```python
district.shape
district.describe()
```

```python
district.to_csv("district_cleaned.csv", index=False)
```

```python
important_cols = [
    "AC_YEAR",
    "STATNAME",
    "DISTNAME",
    "TOTPOPULAT",
    "OVERALL_LI",
    "FEMALE_LIT",
    "MALE_LIT",
    "SCHTOT",
    "ENRTOT",
    "TCHTOT"
]

edu_summary = district[important_cols]
edu_summary.head()
```

```python
edu_summary["STR"] = edu_summary["ENRTOT"] / edu_summary["TCHTOT"]
```

```python
edu_summary["SCHOOLS_PER_10K"] = (edu_summary["SCHTOT"] / edu_summary["TOTPOPULAT"]) * 10000
```

```python
state_summary = edu_summary.groupby("STATNAME").agg({
    "TOTPOPULAT": "sum",
    "SCHTOT": "sum",
    "ENRTOT": "sum",
    "TCHTOT": "sum"
}).reset_index()

state_summary["STR"] = state_summary["ENRTOT"] / state_summary["TCHTOT"]
```

```python
edu_summary.to_csv("district_summary.csv", index=False)
state_summary.to_csv("state_summary.csv", index=False)
```
